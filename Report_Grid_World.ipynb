# Grid World Demo for Value Iteration

This notebook gives a working example of my value iteration implementation for a small finite Markov Decision Process (MDP). The example used here is the grid world problem given in the coursework.

## Basic structure of the code base

The code base is organised into a few small modules. The `mdp.py` file defines the main MDP data structure and the transition object. The `policy.py` file contains functions related to action values and greedy policy extraction. The `solver.py` file contains the main value iteration algorithm.

The purpose of this structure is to keep the main ideas separate. The MDP specification, the Bellman calculations, and the main iterative solver are related, but they do not all need to be written in one file. I think this makes the code easier to read and easier to test.

In this notebook, I only use the finished package and apply it to the coursework grid world. So this notebook acts as a simple demonstration of how the package is intended to be used.

## Modelling choices

I define the grid world directly in the notebook using Python variables for states, actions, transitions, and terminal states. I then create one `FiniteMDP` object from these variables.

I chose to write the grid world in this direct way because the problem is very small and the transition table is already given explicitly in the coursework. For this notebook, this is clearer than hiding the example inside a separate helper function.

```python
from src.value_iteration.mdp import FiniteMDP, Transition
from src.value_iteration.solver import value_iteration

# Discount factor used in this notebook
gamma = 0.9

# States in the coursework grid world
states = ["TL", "TR", "BL", "BR"]

# Available actions in each state
actions = {
    "TL": ["R", "D"],
    "TR": ["L", "D"],
    "BL": ["R", "U"],
    "BR": [],
}

# Transition structure from the coursework table
transitions = {
    "TL": {
        "R": [
            Transition(next_state="TR", probability=0.9, reward=-1.0),
            Transition(next_state="BL", probability=0.1, reward=-2.0),
        ],
        "D": [
            Transition(next_state="BL", probability=0.9, reward=-2.0),
            Transition(next_state="TR", probability=0.1, reward=-1.0),
        ],
    },
    "TR": {
        "L": [
            Transition(next_state="TL", probability=0.9, reward=-1.5),
            Transition(next_state="BR", probability=0.1, reward=10.0),
        ],
        "D": [
            Transition(next_state="BR", probability=0.8, reward=15.0),
            Transition(next_state="TL", probability=0.2, reward=-1.0),
        ],
    },
    "BL": {
        "R": [
            Transition(next_state="BR", probability=0.9, reward=20.0),
            Transition(next_state="TL", probability=0.1, reward=-2.5),
        ],
        "U": [
            Transition(next_state="TL", probability=0.8, reward=-0.5),
            Transition(next_state="BR", probability=0.2, reward=5.0),
        ],
    },
}

# Terminal state
terminal_states = {"BR"}

# Final MDP object
grid_world_mdp = FiniteMDP(
    states=states,
    actions=actions,
    transitions=transitions,
    gamma=gamma,
    terminal_states=terminal_states,
)
```

```python
result = value_iteration(
    grid_world_mdp,
    epsilon=1e-10,
    max_iterations=10000
)
```

```python
print("Final value function:")
for state in states:
    print(f"V({state}) = {result.values[state]:.6f}")
```

```python
print("Greedy policy:")
for state in states:
    print(f"pi({state}) = {result.policy[state]}")
```

```python
print("Convergence information:")
print(f"Number of iterations: {result.iterations}")
print(f"Final delta: {result.final_delta:.12f}")
```

## Comments on the result

The output above shows the final state values produced by value iteration and the greedy policy extracted from those values. The terminal state `BR` has value `0` in this implementation, because terminal states are fixed and have no available actions.

This example shows that the code base can represent a small finite MDP, run value iteration on it, and return both a value function and a policy. This is the main purpose of the notebook.
